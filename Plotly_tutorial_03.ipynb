```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
```

```python
df.head()
```

```python
px.histogram(df, x = 'total_bill', color = 'time', nbins = 40, histnorm= 'probability density', marginal = 'box', title = 'Restaurante', color_discrete_sequence= ['green', 'orange'])
```
